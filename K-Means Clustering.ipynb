```python
def update_assignments(data, centroids):
    c = []
    for i in data:
        c.append(np.argmin(np.sum((i.reshape((1, 784)) - centroids) ** 2, axis=1)))
    return c
```

```python
def update_centroids(data, num_clusters, assignments):
    cen = []
    for c in range(len(num_clusters)):
        cen.append(np.mean([data[x] for x in range(len(data)) if assignments[x] == c], axis=0))
    return cen
```

```python
centroids = (np.random.normal(size=(5, 784)) * 0.0001) + np.mean(x1, axis=0).reshape((1, 784))
```

```python
for i in range(100):
    a = update_assignments(x1, centroids)
    centroids = update_centroids(x1, centroids, a)
    centroids = np.array(centroids)
```
