```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
test = pd.read_csv("C:/Users/adith/Desktop/Datasets/Linear_test.csv")
test = (test - np.min(test)) / (np.max(test) - np.min(test)) 

train = pd.read_csv("C:/Users/adith/Desktop/Datasets/Linear_train.csv")
train = (train - np.min(train)) / (np.max(train) - np.min(train)) 
train.head()
```

```python
X = train.values[:, :-1]
Y = train.values[:, -1].reshape(train.values.shape[0], 1)

X_test = test.values[:, :-1]
Y_test = test.values[:, -1].reshape(test.values.shape[0], 1)

print(X.shape)
print(Y.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
def model(X, Y, learning_rate, iteration):
    m = Y.size
    theta = np.zeros((21,1))
    cost_list = []
    
    for i in range(iteration):
        y_pred = np.dot(X, theta)
        cost = (1/(2*m))*np.sum(np.square(y_pred - Y))
        
        d_theta = (1/m)*np.dot(X.T, y_pred - Y)
        theta = theta - learning_rate*d_theta
        
        cost_list.append(cost)
        
        if(i%(iteration/10) == 0):
            print('Cost is:',cost)
            
    return theta, cost_list
```

```python
iteration = 1000
learning_rate = 0.3
theta, cost_list = model(X, Y, learning_rate = learning_rate, iteration = iteration)
```

```python
y_pred = np.dot(X_test, theta)
error = (1/X_test.shape[0])*np.sum(np.abs(y_pred - Y_test))
rmse = np.sqrt(((y_pred-Y_test)**2).mean())
print(rmse)
print("Test error is:", error*100,"%")
print("Test accuracy is:", (1-error)*100, "%")
```
